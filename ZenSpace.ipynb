```python
import gradio as gr
import pandas as pd
import pickle
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder

# Load the pre-trained random forest model
rf_model =pickle.load(open('finalized_model.sav', 'rb'))

# Load the pre-trained LabelEncoder 
from sklearn.preprocessing import LabelEncoder
import pickle
categ = ['sleep', 'diet', 'care', 'freq', 'meds', 'coun', 'act', 'break',
        'journ', 'social', 'goals', 'self', 'symp', 'gender', 'cry', 'profession', 'field', 'injury', 'negative']

le_sleep=pickle.load(open('label_encoder1.sav', 'rb'))
le_diet=pickle.load(open('label_encoder2.sav', 'rb'))
le_care=pickle.load(open('label_encoder3.sav', 'rb'))
le_freq=pickle.load(open('label_encoder4.sav', 'rb'))
le_meds=pickle.load(open('label_encoder5.sav', 'rb'))
le_coun=pickle.load(open('label_encoder6.sav', 'rb'))
le_act=pickle.load(open('label_encoder7.sav', 'rb'))
le_break=pickle.load(open('label_encoder8.sav', 'rb'))
le_journ=pickle.load(open('label_encoder9.sav', 'rb'))
le_social=pickle.load(open('label_encoder10.sav', 'rb'))
le_goals=pickle.load(open('label_encoder11.sav', 'rb'))
le_self=pickle.load(open('label_encoder12.sav', 'rb'))
le_symp=pickle.load(open('label_encoder13.sav', 'rb'))
le_gender=pickle.load(open('label_encoder14.sav', 'rb'))
le_cry=pickle.load(open('label_encoder15.sav', 'rb'))
le_profession=pickle.load(open('label_encoder16.sav', 'rb'))
le_field=pickle.load(open('label_encoder17.sav', 'rb'))
le_injury=pickle.load(open('label_encoder18.sav', 'rb'))
le_negative=pickle.load(open('label_encoder19.sav', 'rb'))




# Define the input questions for the model
q1='How many hours of sleep did you get last night?'
q2='Have you been eating regularly and maintaining a healthy diet?'
q3='Have you been practicing self-care regularly?'
q4='If so how regularly?'
q5='Have you been taking any medication as prescribed?'
q6='Have you been attending therapy or counseling sessions regularly?'
q7='Have you been engaging in physical activity or exercise regularly?'
q8='Have you been taking breaks and practicing relaxation techniques throughout the day?'
q9='Have you been journaling or practicing mindfulness?'
q10='Have you been socializing and connecting with others regularly?'
q11='Have you been setting and achieving personal goals?'
q12='Have you been practicing positive self-talk and challenging negative thoughts?'
q13='Have you been experiencing any physical symptoms related to your mental health, such as headaches or stomach problems?'
q14='How old are you?'
q15='What is your Gender ?'
q16='rate how healthy your diet is on a scale of 1-10'
q17='How many times did you cry this week?'
q18='What is your profession? - profession'
q19='Do you like the field your in?'
q20='Have you ever been in any situation in which you were seriously injured or feared that you were seriously injured?'
q21='In past week, have you had any negative thoughts or thoughts about self-harming?'

# Define the function for making predictions using the random forest model
def classify_answers(Q1, Q2, Q3, Q4, Q5, Q6, Q7, Q8, Q9, Q10, Q11, Q12, Q13, Q14, Q15, Q16, Q17, Q18, Q19, Q20, Q21):
    
    # Create a dataframe from the input questions
    answers = pd.DataFrame(data={'sleep': [Q1],
                            'diet': [Q2],
                            'care': [Q3],
                            'freq': [Q4],
                            'meds': [Q5],
                            'coun': [Q6],
                            'act': [Q7],
                            'break': [Q8],
                            'journ': [Q9],
                            'social': [Q10],
                            'goals': [Q11],
                            'self': [Q12],
                            'symp': [Q13],
                            'age': [Q14],
                            'gender': [Q15],
                            'dietrate': [Q16],
                            'cry': [Q17],
                            'profession': [Q18],
                            'field': [Q19],
                            'injury': [Q20],
                            'negative': [Q21]})
    
    
    # Encode Categorical Columns
   
    answers['sleep']=le_sleep.transform(answers['sleep'])
    answers['diet']=le_diet.transform(answers['diet'])
    answers['care']=le_care.transform(answers['care'])
    answers['freq']=le_freq.transform(answers['freq'])
    answers['meds']=le_meds.transform(answers['meds'])
    answers['coun']=le_coun.transform(answers['coun'])
    answers['act']=le_act.transform(answers['act'])
    answers['break']=le_break.transform(answers['break'])
    answers['journ']=le_journ.transform(answers['journ'])
    answers['social']=le_social.transform(answers['social'])
    answers['goals']=le_goals.transform(answers['goals'])
    answers['self']=le_self.transform(answers['self'])
    answers['symp']=le_symp.transform(answers['symp'])
    answers['gender']=le_gender.transform(answers['gender'])
    answers['cry']=le_cry.transform(answers['cry'])
    answers['profession']=le_profession.transform(answers['profession'])
    answers['field']=le_field.transform(answers['field'])
    answers['injury']=le_injury.transform(answers['injury'])
    answers['negative']=le_negative.transform(answers['negative'])

    answers = answers.apply(pd.to_numeric)
    
    
    prediction = rf_model.predict(answers)
    
    
    # Return the predicted class
    if prediction[0] == 0:
        return "According to your answers, You can try some fun tasks to feel better"
    
    
    elif prediction[0] == 1:
        return "According to your answers, You seem to be in a good mental state :)"
   
    else:
        return "According to your answers, We reccomend that you talk to a therapist"

# Define the input interface for Gradio
inputs = [gr.Dropdown(
            ['0-3','4-6','6-8','10-12','More than 12'], label=q1, info="Choose one"),
          gr.Dropdown(
            ['Yes','No'], label=q2, info="Choose one"),
          gr.Dropdown(
            ['Yes','No'], label=q3, info="Choose one"),
          gr.Dropdown(
            ['Not at all','Everyday','4-5 times a week','2-3 times a week'], label=q4, info="Choose one"),
          gr.Dropdown(
            ['Yes','No'], label=q5, info="Choose one"),
          gr.Dropdown(
            ['Yes','No'], label=q6, info="Choose one"),
          gr.Dropdown(
            ['Yes','No'], label=q7, info="Choose one"),
          gr.Dropdown(
           ['Yes','No'], label=q8, info="Choose one"),
          gr.Dropdown(
           ['Yes','No'], label=q9, info="Choose one"),
          gr.Dropdown(
           ['Once a week','2-4 times a week','5 and more'], label=q10, info="Choose one"),
          gr.Dropdown(
           ['Yes','No'], label=q11, info="Choose one"),
          gr.Dropdown(
           ['Yes','No'], label=q12, info="Choose one"),
          gr.Dropdown(
           ['Yes','No'], label=q13, info="Choose one"),
          gr.Slider(1, 100, value=20, label=q14, info="Choose betwen 1 and 100"),
          gr.Dropdown(
           ['Male','Female','Other','Rather Not say'], label=q15, info="Choose one"),
           gr.Slider(1, 10, value=1, label=q16, info="Choose between 1 and 10, 10 being very healthy"),
          gr.Dropdown(
           ['Not at all','1-2 times','3-4 times'], label=q17, info="Choose one"),
          gr.Dropdown(
           ['School Student','College Student','Working Professional','Other'], label=q18, info="Choose one"),
          gr.Dropdown(
           ['Yes','No'], label=q19, info="Choose one"),
          gr.Dropdown(
           ['Yes','No','Maybe'], label=q20, info="Choose one"),
          gr.Dropdown(
          ['1-3', 'Not at all', 'More than 5','4-5'], label=q21, info="Choose one"),
         ]

# Define the output interface for Gradio
output = gr.outputs.Textbox(label="Our Prediction is: ")



#https://wallpapers.com/images/featured/nsup0okwyoqg4db9.jpg
# Run the Gradio interface
gr.Interface(fn=classify_answers, inputs=inputs, outputs=output, title="ZenSpace Mental Health Detector",theme=gr.themes.Base(),
         allow_flagging='never').launch(share=True)
```
